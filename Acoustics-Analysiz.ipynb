Import library

```python
from scikits.audiolab import wavread, wavwrite
import matplotlib.pyplot as plt
import numpy as np
```

```python
data, fs, enc = wavread('sounds/Snap.wav')
plt.plot(data[0:300, 0])
plt.grid(1)
plt.ylabel('Voltage')
plt.xlabel('Samples')
plt.show()
```

```python
from IPython.display import Audio
Audio('sounds/Snap.wav')
```

```python
def inver(x):
    if x < 0:
        return -x
    else:
        return x
data2 = map(inver,data[0:300,0])
```

```python
plt.plot(data2[0:300])
plt.grid(1)
plt.ylabel('Voltage')
plt.xlabel('Samples')
plt.show()
```

```python
alpha = 0.99
data3 = np.zeros(300)
for i in range(len(data2)):
    data3[i] = data2[i] * (1 - alpha) + alpha * data3[i-1]
```

```python
plt.plot(data3[0:300])
plt.grid(1)
plt.ylabel('Voltage')
plt.xlabel('Samples')
plt.show()
```

```python
k = 10 
data4 = np.zeros(300)
for i in range(len(data2)):
    data4[i] = data3[i - i%k]
```

```python
plt.plot(data4[0:300])
plt.grid(1)
plt.ylabel('Voltage')
plt.xlabel('Samples')
plt.show()
```

```python
data5 = data4 - np.mean(data4) * np.ones(300)
```

```python
plt.plot(data5[0:300])
plt.grid(1)
plt.ylabel('Voltage')
plt.xlabel('Samples')
plt.show()
```
